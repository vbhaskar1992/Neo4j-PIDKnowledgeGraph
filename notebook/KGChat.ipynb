```python
from langchain_openai import AzureChatOpenAI
```

```python
def create_llm():
    llm = AzureChatOpenAI(
        api_key= azure_api_key,
        azure_endpoint= azure_endpoint,
        azure_deployment= azure_deployment,
        api_version=azure_api_version
       # max_tokens= 800
)
    return llm
```

```python
llm = create_llm()
```
